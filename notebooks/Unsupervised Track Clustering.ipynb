Unsupervised Track Clustering

In this part of our project, we focus on clustering Formula 1 race tracks by their
characteristics, using the Ergast API for data. Features such as track layout, turn
complexity, and historical race data like average lap times and overtaking
frequencies are pivotal. This approach shifts the analytical spotlight from drivers
and teams to the tracks, offering insights into how different racing environments
influence strategies and outcomes. Each track's unique configuration poses distinct
challenges, enhancing our understanding of the sport’s dynamics. For clustering, we
choose K-Means for its efficiency with large datasets, DBSCAN for its ability to
handle diverse cluster shapes and detect outliers, and Hierarchical Clustering for
revealing track similarities at various levels. These algorithms are selected for their
complementary strengths in handling the complex and varied nature of F1 race
tracks.
Our methodology involves meticulous data preparation, including cleaning and
feature engineering, to accurately reflect each track's characteristics. Physical
attributes like track length and number of turns, alongside performance indicators
such as overtaking possibilities, are integral to our comprehensive analysis.
Evaluating the clustering will initially utilize the Silhouette score, with openness to
other methods that might provide additional insights. Visualization will be a
dynamic process, potentially employing cluster maps and heatmaps, depending on
their effectiveness in illustrating our results. This flexible and in-depth exploration
aims to deepen the understanding of F1 racing, highlighting the significant role of
race tracks in shaping the sport's complexity and excitement. Our findings are
anticipated to offer valuable perspectives for fans, teams, and strategists,
showcasing the intricate connection between track design and race strategy
