# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

## Circuits Visualizations

```python
circuits = pd.read_csv('../data/raw/circuits.csv')
races = pd.read_csv('../data/raw/race_results.csv')
```

```python
m = folium.Map(location=[circuits.lat.mean(), circuits.lng.mean()], zoom_start=2)

for i, loc in circuits.iterrows():
    name = loc.name
    folium.Marker([loc.lat, loc.lng]).add_to(m)
m
```
